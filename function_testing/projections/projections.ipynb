```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PLOTS = Path().cwd().parent / 'plots'

# Font size of plots
plt.rcParams['font.size'] = 18

FIG_SIZE = (20, 10)
```

```python
edge = pd.read_csv('edge_building_projections.csv',
                   usecols = ['year', 'variable', 'value', 'unit'],
                   index_col = ['variable', 'year'])

gcam = pd.read_csv('gcam_building_projections.csv',
                   usecols = ['year', 'variable', 'value', 'unit'],
                   index_col = ['year', 'variable'])
```

## Comparison Plots

### Floorspace

```python
edge.loc['res_floor_proj', 'value'] = edge.loc['res_floor_proj', 'value'].copy().multiply(2)
```

```python
fig, ax = plt.subplots(figsize = FIG_SIZE)
edge[edge['variable'] == 'res_floor_proj'].rename(columns = {'value': 'Residential (EDGE)'}).plot(ax = ax,
                                                                                           color = 'orange')
edge[edge['variable'] == 'com_floor_proj'].rename(columns = {'value': 'Commercial (EDGE)'}).plot(ax = ax,
                                                                                          color = 'orange',
                                                                                          ls = '--')
gcam[gcam['variable'] == 'res_floor_proj'].rename(columns = {'value': 'Residential (GCAM)'}).plot(ax = ax,
                                                                                           color = 'red')
gcam[gcam['variable'] == 'com_floor_proj'].rename(columns = {'value': 'Commercial (GCAM)'}).plot(ax = ax,
                                                                                          color = 'red',
                                                                                          ls = '--')
ax.legend()
ax.set(xlabel = 'Year',
       ylabel = 'Floor space demand (m$^2$/cap)');
fig.savefig(PLOTS / 'floor_demand_comparison.png')
```
